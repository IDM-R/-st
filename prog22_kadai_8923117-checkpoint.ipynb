## に基づいて，0 ステップ目から <span style="color: blue; ">200 ステップ</span>目までの位置に関する次の２点の情報：<br><br>（１）位置の移動経過＜折れ線グラフ＞<br>（２）ステップ毎にいる位置の頻度＜ヒストグラム，例 10.1（テキスト p.125）＞<br><br>を図示せよ．必要に応じて，複数の Code セルを用いて良い．

```python
# 問１の入力欄
%matplotlib inline
import matplotlib.pyplot as plt

def next_step():
    if random.random() < .6:
        return 1
    else:
        return -1

pos = 0
pos_history = [0] 

for i in range(200):
    pos += next_step()
    pos_history.append(pos)

plt.plot(pos_history)
plt.title('Position for each step',{"fontsize":15})
plt.xlabel('step')
plt.ylabel('position')
plt.show()
```

```python
plt.hist(pos_history, rwidth=0.8) 
plt.title('Frequency of positions',{"fontsize":15})
plt.xlabel('position')
plt.show()
```
