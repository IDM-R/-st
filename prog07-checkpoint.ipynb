## 4.1 while 文を使った繰り返し（p.43～）

## while 文は与えられた条件式が成り立つ間はずっと，繰り返して命令を実行する．Python の while 文は次の書式をとる．

```python
while 条件式:
    実行する命令1
    実行する命令2
```

## while 文の記述では次の３点に注意する．<br><br>(1) if 文と同じく，条件式の後に半角コロン「:」が必要<br>(2) 実行する命令は必ず段付け（インデント）を与えて記述する<br>(3) 段付けされている行は，全て繰り返し実行される

## 例１．1 から 7 までの数字を表示するプログラム

```python
i = 1 # 変数 i を使い，初期値として i を 1 と設定する
while i <= 7: # i が 7 より大きくなったら，while 文から脱出し，繰り返し処理は終了される
    print(i) 
    i = i + 1 # i のカウントアップ
```

## 上のプログラムにおける変数 i は，ループカウンタと呼ばれる．また，2.4.4 節の複合演算子の表（表2.8）の１行目にあるように，i のカウントアップに対応する文 i = i + 1 を，<span style="color: blue; ">i += 1</span>という文に置き換えても良い．<br><br>例１（別解）．1 から 7 までの数字を表示するプログラム

```python
i = 1 # 変数 i を使い，初期値として i を 1 と設定する
while i <= 7: # i が 7 より大きくなったら，while 文から脱出し，繰り返し処理は終了される
    print(i) 
    i += 1 # i のカウントアップ
```

## 例２．文字列「Hello World」を５回繰り返し表示するプログラム

```python
i = 1 # 変数 i を使い，初期値として i を 1 と設定する
while i <= 5: # i が 5 より大きくなったら，while 文から脱出し，繰り返し処理は終了される
    print("Hello World") 
    i += 1 # i のカウントアップ
```

## プログラム作成の際に，条件式を間違えたり，変数 i をカウントアップさせる行を忘れると，処理が永遠に終わらない．この状態を「無限ループ」と呼ぶ．<br><br>例３．例２のプログラムにおいて，４行目の変数 i のカウントアップを忘れた場合（要注意．実行してすぐに無限ループが現れる）

```python
i = 1 # 変数 i を使い，初期値として i を 1 と設定する
while i <= 5: # i が 5 より大きくなったら，while 文から脱出し，繰り返し処理は終了される
    print("Hello World")
```

## なお，最初から条件を満たさないプログラムは，一度もループせずにそのまま処理は終了する．

```python
i = 1 # 変数 i を使い，初期値として i を 1 と設定する
while i > 5: # そもそも初期値 1 が代入された i は i>5 を始めから満たしていないため，繰り返し処理は何も行われずに終了される
    print("Hello World") 
    i += 1
```

## 例４．while 文を使った，1 から 20 までの数の合計を求めるプログラム

```python
i = 1
s = 0 # 合計値を格納するための文字を s として，初期値を 0 とする
while i <= 20:
    s += i # 1 から 20 までの足し上げ
    i += 1 # 変数 i のカウントアップ
print("求める合計値は",s) # ループを抜けた後の最終的な合計値の出力
```

## 例５．while 文を使った，１０個の数 $1^4,2^4,\ldots,10^4$ の合計を求めるプログラム

```python
i = 1
s = 0 # 合計値を格納するための文字を s として，初期値を 0 とする
while i <= 10:
    s += i**4 # 足し上げ
    i += 1    # 変数 i のカウントアップ
print("求める合計値は",s)
```

## 4.2 for 文を使った繰り返し（p.46～）

## 繰り返し実行を行う別の方法として，for 文を使うこともできる．for 文では，繰り返したい要素を与えると，その要素の数だけループを繰り返す．Python の for 文は次の書式をとる．

```python
for 変数 in シーケンス:
    実行する命令1
    実行する命令2
```

## for 文の記述では次の５点に注意する．<br><br><span style="color: blue; ">(1) 繰り返しで使用する「変数」を指定する </span><br><span style="color: blue; ">(2) 変数とシーケンスの間には，キーワード「in」を挿入する</span><br>(3) if 文や while 文と同様に，シーケンスの後に半角コロン「:」が必要<br>(4) 実行する命令は必ず段付け（インデント）を与えて記述する<br>(5) 段付けされている行は，全て繰り返し実行される

## 例６－１．for 文を使った，1 から 7 までの数字それぞれを４倍した数を順に（数字の小さい順に）印字するプログラム

```python
for i in [1,2,3,4,5,6,7]: # 変数 i に，1, 2, 3, 4, 5, 6, 7 を順次代入し，以下を実行
    print(4*i)
```

## 例６－２．for 文を使った，1 から 7 までの数字それぞれを４倍した数を逆順に（数字の大きい順に）印字するプログラム<br><br>注意．要素の並び順が例６－１とは異なるので，変数 i に代入される数字の順も異なる．そのため，出力される４倍された数字の順も同じとはならない．

```python
for i in [7,6,5,4,3,2,1]: # 変数 i に，7, 6, 5, 4, 3, 2, 1 を順次代入し，以下を実行
    print(4*i)
```

## シーケンスの中身は，例６の１行目終わりのような数字でなくても構わない．

```python
for i in ["January","February","March"]:
    print("今月は",i,"です")
```

## 例７．list() によるシーケンスの中身の確認

```python
# 4 から 7 までのシーケンスを生成し，そのシーケンスを a とおく
a = range(4,8)
# list() を使ってシーケンスの中身を確認する
list(a)
```

## 注意．表4.1の赤字部分で書かれているように，range() の生成するシーケンスは，２番目の引数を含まない！

```python
for i in range(4,8): # 2 番目の引数となる 8 ではなく，8 の一つ手前の数 7 までがシーケンスに属する
    print(i)
```

## range() の３番目の引数を使うと，生成する数値の増分を指定することができる．

```python
for i in range(-20,60,10):
    print(i)
```

## 例８．for 文と range 関数を使った，1 から 20 までを逆順に（数字の大きい順に）足し上げて合計を求めるプログラム

```python
s = 0
for i in range(20,0,-1):
    s += i
print("求める合計値は",s)
```

## 例９．for 文と range 関数を使った，1 から 30 までの偶数を逆順に（数字の大きい順に）足し上げて合計を求めるプログラム

```python
s = 0
for i in range(30,0,-2):
    s += i
print("求める合計値は",s)
```
