## 問１－１．講義ノートの例１－２に倣って，$$a_n:=7^n+2 \ (n \geq 0, n \in \mathbb{Z})$$ の再帰的定義を与えなさい（下記の再帰的定義を記述した <span style="color: blue; ">Markdown セルをダブルクリックし，３ヶ所の "..." に適切な数字を入力し，Shift キー＆Enter キーを再度押して</span>，元の表示に戻すこと）．

$$\left\{ \begin{array}{l} a_0=3,\\a_n=7*a_{n-1}-12 \ \ (n \geq 1) \end{array} \right.$$

## 問１－２．<span style="color: blue; ">再帰的定義により</span> $7^n+2 \ (n \geq 0, n \in \mathbb{Z})$ を計算する関数 ex7() を作りなさい．また，$n$ を 0 から 5 まで 1 ずつ大きくしたときの ex7($n$) の値を表示するプログラムを <span style="color: blue; ">for 文を用いて</span>作りなさい．

```python
# 問１－２の入力欄：再帰的定義により，関数 ex7() を定義するためのプログラム
def ex7(n):
    if n == 0:
        return 3          
    else:
        return 7*ex7(n-1)-12 
```

```python
# 問１－２の入力欄：n を 0 から 5 まで 1 ずつ大きくしたときの ex7(n) の値を表示する「for 文を用いた」プログラム
for n in range(0,6):
    print(ex7(n))
```

## 問２．次の再帰的定義：

$$\left\{ \begin{array}{l} a_0=3, \ a_1=9,\\a_n=49 a_{n-2}-96 \ \ (n \geq 2) \end{array} \right.$$

## によって定まる $a_n \ (n \geq 0, n \in \mathbb{Z})$ を計算する関数 ex49() を作りなさい．また，$n$ を 0 から 5 まで 1 ずつ大きくしたときの ex49($n$) の値を表示するプログラムを <span style="color: blue; ">for 文を用いて</span>作りなさい．

```python
# 問２の入力欄：再帰的定義により，関数 ex49() を定義するためのプログラム
def ex49(n):
    if n==0:
        return 3
    if n==1:
        return 9
    else:
        return 49*ex49(n-2)-96
```

```python
# 問２の入力欄：n を 0 から 5 まで 1 ずつ大きくしたときの ex49(n) の値を表示する「for 文を用いた」プログラム
for n in range(0,6):
    print(ex49(n))
```

## 問３（発展）．折れ線グラフにより，$n=0,1,...,5$ に対する関数 ex7($n$) の値をプロットするプログラムを作りなさい．

```python
# 問３の入力欄
%matplotlib inline
import matplotlib.pyplot as plt

def ex7(n):
    if n == 0:
        return 3          
    else:
        return 7*ex7(n-1)-12 
  
n=list(range(0,6)) 
a_n=list(map(ex7, n)) 

plt.plot(n, a_n)
```

## ＜plot() によるグラフ描画（第７章：p.86～，詳しくは後期第１回でふれる）＞<br><br>　%matplotlib inline：<br>　Jupyter Notebook 上で同一ウインドウ内にグラフを描画するための宣言<br>　（注意．"%" と "matplotlib" の間に，空白スペースを挟んではいけない！）<br><br>　import matplotlib.pyplot as plt：<br>　matplot の pyplot モジュールを plt（別名）としてインポートするための命令文<br>　（注意．"plt" の部分は他の文字でも構わないが，通例として "plt" が用いられる！）<br><br>例１．折れ線グラフにより，リスト内の要素の組を表示するプログラム

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [100, 200, 300, 400, 500, 600]
y = [10, -20, 30, -50, 80, -130]

plt.plot(x, y) # リスト x とリスト y の同成分の要素の組をプロットする命令文
```

## 例２．折れ線グラフにより，$x=-10,-9,...,10$ に対する関数 $f(x)=x^3$ の値をプロットするプログラム

```python
%matplotlib inline
import matplotlib.pyplot as plt

def f(x):
    return x**3

x=list(range(-10,11)) # x の範囲の右端の値＝関数 range() の第２引数から「1 を引いた」値
y=list(map(f, x))     # map(f,x)：リスト x の各要素に与えられた関数 f を適用し，その結果を集めたオブジェクトを返す命令文
plt.plot(x, y)
```
