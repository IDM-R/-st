## 問１－１．<span style="color: blue; ">for 文の２重ループを用いて</span>，次の数：$$\sum_{i=1}^n \sum_{j=1}^n \frac{i}{j+2^i}$$ を計算する関数 f($n$) を作成し，f(100) を表示しなさい．

```python
# 問１－１の入力欄：関数 f() を定義するためのプログラム
def f(n):
    s=0
    list1=range(1,n+1) 
    
    for i in list1: 
        for j in list1:
            s += i/(j+2**i)
            
    return s
    
```

```python
# 問１－１の入力欄：f(100) を表示するためのプログラム
print ("f(100)=",f(100))
```

## 問１－２．<span style="color: blue; ">time モジュールを使って</span> f($n$) を計算するための実行時間が <span style="color: blue; ">0.03 秒以上 1 秒未満である</span>ような自然数 $n$ を求め，そのときの実行時間を表示しなさい．<br><br>【注意】講義（第１４回）ノートの例２と同様に，所定の計測範囲に収まったときの $n$ の値を<span style="color: blue; ">１つだけ</span>選んで，$n$ の値とそのときの実行時間の２点を表示させる形で構いません．

```python
# 問１－２の入力欄
import time
for n in range(200,300) :
    t1 = time.time() 
    f(n)          
    t2 = time.time()
    s=t2-t1
    if 0.03<=s<1.00:
        print(n,t1,t2,s)
```

## 問２－１．<span style="color: blue; ">for 文の３重ループを用いて</span>，次の数：$$\sum_{i=1}^n \sum_{j=1}^n \sum_{k=1}^n  \frac{1}{i+2j+3k}$$ を計算する関数 g($n$) を作成し，g(100) を表示しなさい．

```python
# 問２－１の入力欄：関数 g() を定義するためのプログラム
def g(n):
    s=0
    list1=range(1,n+1) 
    
    for i in list1: 
        for j in list1:
            for k in list1:
                s += 1/(i+2*j+3*k)
            
    return s
```

```python
# 問２－１の入力欄：g(100) を表示するためのプログラム
print ("g(100)=",g(100))
```

```python
# 問２－２の入力欄
import time
for n in range(220,221) :
    t1 = time.time() 
    g(n)          
    t2 = time.time()
    s=t2-t1
    if 3<=s<10:
        print("n="n,t1,t2,s)
```
