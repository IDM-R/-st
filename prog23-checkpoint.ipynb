## 上記の画像における確率的な決め方の規則②に基づいて，各ステップでの位置を決める関数 next_step2() を次に示す．ただし，移動方向と $x$ 軸の正の向きとの成す角 $\theta$ を，プログラム内では th とする．また，プログラムの最終行のように，return 文で複数の値を指定することで，関数から複数の値を返すことができる．

```python
import math
def next_step2():
    th = random.random() * math.pi * 2 # th：0 以上 2π 未満の乱数
    return math.cos(th), math.sin(th) # 第１成分：x 軸に沿った移動距離，第２成分：y 軸に沿った移動距離
```

## 【確認】next_step2() を 1 回適用して 1 ステップ目の位置を計算し，表示する．ただし，初期（0 ステップ目）の位置座標を (0,0) として，位置の座標を (posx,posy) とする（以降，実行の度に出力が変わる）．

```python
posx = 0 # 初期位置
posy = 0 # 初期位置

dx, dy = next_step2()
posx = posx + dx
posy = posy + dy
print("posx", posx, "posy", posy)
```

## 上記の確認で問題は見られなかったため，next_step() を 5 回適用して 0 ステップ目から 5 ステップ目までの位置を計算し，表示する．ただし，位置を pos として，初期（0 ステップ目）の位置を 0 とする．

```python
posx = 0 # 初期位置
posy = 0 # 初期位置
print("posx", posx, "posy", posy,"：初期位置") # 初期位置の表示

for i in range(5):
    dx, dy = next_step2()
    posx += dx # x 軸上の変位 dx を現在の位置に加えて，次の位置を計算
    posy += dy # y 軸上の変位 dy を現在の位置に加えて，次の位置を計算
    print("posx", '{:.3f}'.format(posx), "posy", '{:.3f}'.format(posy)) # '{:.3f}'.format(num)：num の小数点以下３桁を表示
```

## 次に，matplotlib を使って，規則②に基づいた 0 ステップ目から 100 ステップ目までの位置に関する次の２点の情報：<br><br>（１）位置の移動経過＜折れ線グラフ＞<br>（２）ステップ毎にいる位置の頻度＜ヒストグラム，例 10.1（テキスト p.125）＞<br><br>を図示する．

```python
%matplotlib inline
import matplotlib.pyplot as plt

posx = 0
posy = 0
posx_history = [0] # posx_history：ステップ毎にいる位置の x 座標を記録するリスト（初期の位置の x 座標のみ格納済み）
posy_history = [0] # posy_history：ステップ毎にいる位置の y 座標を記録するリスト（初期の位置の y 座標のみ格納済み）

for i in range(100):
    dx, dy = next_step2()
    posx += dx
    posy += dy
    posx_history.append(posx) # 位置の x 座標を記録
    posy_history.append(posy) # 位置の y 座標を記録

plt.plot(posx_history, posy_history, marker='o', markersize=5) # 各ステップでの位置を丸で表示
plt.plot(0, 0, marker='.', markersize=20) # 初期位置 (0,0) を点で表示
plt.plot(posx_history[-1], posy_history[-1], marker='s', markersize=10) # 最後のステップでの位置を四角形で表示

plt.title('Position for each step',{"fontsize":15})
plt.xlabel('$x$-axis')
plt.ylabel('$y$-axis')
plt.show() # グラフ表示
```

## 続いて，始点から終点までの位置変化を追跡したい．そこで，animation モジュールをインポートすることで，ステップ毎の位置を動的に観察できる．

```python
# 動くグラフを表示するための文
%matplotlib notebook

import matplotlib.pyplot as plt
import matplotlib.animation as animation # animation モジュールをインポート

fig = plt.figure()
ims = []

posx = 0
posy = 0
posx_history = [0] # posx_history：ステップ毎にいる位置の x 座標を記録するリスト（初期の位置の x 座標のみ格納済み）
posy_history = [0] # posy_history：ステップ毎にいる位置の y 座標を記録するリスト（初期の位置の y 座標のみ格納済み）

for i in range(100):
    dx, dy = next_step2()
    posx += dx
    posy += dy
    posx_history.append(posx) # 位置の x 座標を記録
    posy_history.append(posy) # 位置の y 座標を記録
    im = plt.plot(posx_history, posy_history, marker='o', color="black", markersize=5) # 各ステップでの位置を丸で表示
    ims.append(im)
    
plt.plot(0, 0, marker='.', markersize=20) # 初期位置 (0,0) を点で表示
plt.plot(posx_history[-1], posy_history[-1], marker='s', markersize=10) # 最後のステップでの位置を四角形で表示

plt.title('Position for each step',{"fontsize":15})
plt.xlabel('$x$-axis')
plt.ylabel('$y$-axis')

# アニメーション実行（連続表示するプロットを 100 ms毎に更新するよう指定している）
ani = animation.ArtistAnimation(fig, ims, interval=100)

plt.show() # グラフ表示
```
