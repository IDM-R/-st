## 問３．講義（後期，第２回）ノートの例１－４にならって，linspace() を使って，$-1 \leq x \leq 1$ における $f_1(x)=x^4-2x+2,f_2(x)=2-5x^3$ の<span style="color: blue; ">滑らかな</span>グラフを描画せよ．

```python
# 問３の入力欄
import numpy as np
def f_1(x):
    return x**4-2*x+2
def f_2(x):
    return 2-5*x**3
x=np.linspace(-1,1)
plt.plot(x,f_1(x))
plt.plot(x,f_2(x))
```
