## 11.1 モンテカルロ法とは（p.128～）<br><br>　ある事象が確率的な振る舞いをするときに，乱数や確率分布を使って解を得ようとするアルゴリズムの総称を「モンテカルロ法」という．今回は，モンテカルロ法を用いて次の２つの実験を行う．<br><br>実験１．円周率 $\pi=3.1415926535...$ の近似値を調べる．<br><br>下図のように，横軸の区間を $[-1,1]$ とした一辺の長さが 2 の正方形に内接する半径 1 の円で囲まれた領域を考える．

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np

### 円の作図 ###
def f1(x):
    return (1-x**2)**(1/2) # 上半分の円の方程式

def f2(x):
    return -(1-x**2)**(1/2) # 下半分の円の方程式

x = np.linspace(-1, 1, 201)

### グラフ描画 ###
plt.plot(x, f1(x), color = "black")
plt.plot(x, f2(x), color = "black")

plt.axis('square') # 表示される図の縦横比を 1:1 にする
plt.xlim(-1, 1)
plt.ylim(-1, 1)

plt.show()
```

## 次に，正方形の中に，ランダムに $N$ 個の点を打つ．領域全体の正方形内に打たれた点が円内にある確率は<br><br>円の面積 $\div$ 正方形の面積 $= \pi \div 4 = \frac{\pi}{4}$ <br><br>となる．実際，点の個数 $N$ を大きくとると，<br><br>円内に打たれた点の数 $N_{\rm in}$  $\div$ 領域全体の正方形内に打たれた全ての点の数 $N$ $\simeq \frac{\pi}{4}$ <br><br>であることが知られている．このことから，次の式：<br><br>$\pi \simeq$（円内に打たれた点の数 $N_{\rm in}$  $\div$ 領域全体の正方形内に打たれた全ての点の数 $N$）$\times$ $4$ <br><br>を考えることで，$\pi$ の近似値を調べることが可能となる．

## 【確認】$N=100000$ の場合

```python
N_in = 0

N = 100000 # 試行回数

for i in range(0,N):

    ran_x = 2 * random.random() - 1 # X の乱数（-1 以上 1 未満）
    ran_y = 2 * random.random() - 1 # Y の乱数（-1 以上 1 未満）

    ran_point = (ran_x**2 + ran_y**2) ** (1/2) # 原点 (0,0) から (ran_x,ran_y) までの距離  

    # 原点 (0,0) から (ran_x,ran_y) までの距離が 1 以下，すなわち，
    # (ran_x,ran_y) が半径 1 の円内にあれば，円領域内に入った点の数 N_in に 1 を加える
    if ran_point <= 1: 
        N_in += 1

Pie = N_in/N*4 # πの近似式

print("円領域内に入った点の数 N_in=", N_in)
print("正方形領域に打たれた全ての点の数 N=", N)

print("実験によって計算されるπの近似値 N_in / N * 4 =", N_in, "/", N, "* 4 =", N_in/N*4)
```

## $\pi$ の近似値を得る上記のプログラムを 8000 回繰り返し，8000 通りで得られる $\pi$ の近似値を下記のプログラムによってヒストグラムで図示する（以降，$N=100$ とする）．

```python
%matplotlib inline
import matplotlib.pyplot as plt

Pie_list = []

M = 8000
N = 100 # 試行回数

for i in range(0,M):
    
    N_in = 0
    for j in range(0,N):

        ran_x = 2*random.random()-1 # X の乱数（-1 以上 1 未満）
        ran_y = 2*random.random()-1 # Y の乱数（-1 以上 1 未満）

        ran_point = (ran_x**2 + ran_y**2) ** (1/2) # 原点 (0,0) から (ran_x,ran_y) までの距離  

        if ran_point <= 1:
            N_in += 1

    Pie = N_in/N*4 # πの近似式
    Pie_list.append(Pie)

### 𝜋 の近似値のばらつきに関するヒストグラム ###
plt.hist(Pie_list, rwidth=0.9)
plt.title('Approx. values of $\pi$',{"fontsize":15})
plt.xlabel('Calculated values')
plt.show() # グラフ表示
```

## 実験２．$\log 2 (=\log_e 2)=0.69314718...$ の近似値を調べる．<br><br>下図のように，横軸の区間を $[1,2]$ とした一辺の長さが 1 の正方形のうち，曲線 $y=\frac{1}{x}$ を境に区切られた領域のうち，下半分の領域を考える．

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np

### 曲線 y=1/x の作図 ###
def f(x):
    return 1/x

x = np.linspace(1, 2, 201)

### グラフ描画 ###
plt.plot(x, f(x), color = "black")

plt.axis('square') # 表示される図の縦横比を 1:1 にする
plt.xlim(1, 2)
plt.ylim(0, 1)

plt.show()
```

## 注意．曲線 $y=\frac{1}{x}$ を境に区切られた領域のうち，下半分の領域の面積を $S$ とおくと，$$S=\int_1^2 \frac{1}{x} \ dx=[ \log x ]_1^2=\log 2-0=\log 2 \ (=\log_e 2)$$ より， $S=\log 2$ である．<br><br>　次に，正方形の中に，ランダムに $N$ 個の点を打つ．領域全体の正方形内に打たれた点が，曲線 $y=\frac{1}{x}$ を境に区切られた領域のうち，下半分の領域内にある確率は<br><br>$S$ $\div$ 正方形の面積 $= \log 2 \div 1 = \log 2$ <br><br>となる．実際，点の個数 $N$ を大きくとると，<br><br>曲線 $y=\frac{1}{x}$ を境に区切られた領域のうち，下半分の領域内に打たれた点の数 $N_{\rm in}$<br>　$\div$ 領域全体の正方形内に打たれた全ての点の数 $N$ $\simeq \log 2$ <br><br>であることが知られている．このことから，次の式：<br><br>$\log 2 \simeq$ 曲線 $y=\frac{1}{x}$ を境に区切られた領域のうち，下半分の領域内に打たれた点の数 $N_{\rm in}$<br>　　　　$\div$ 領域全体の正方形内に打たれた全ての点の数 $N$<br><br>を考えることで，$\log 2$ の近似値を調べることが可能となる．

## 【確認】$N=100000$ の場合

```python
N = 100000 # 試行回数

N_in = 0

for i in range(0,N):

    ran_x = random.random() + 1 # X の乱数（1 以上 2 未満）
    ran_y = random.random() # Y の乱数（0 以上 1 未満）

    if ran_y <= 1/ran_x:
        N_in += 1

Log2 = N_in/N # log 2 の近似式

print("曲線 y=1/x を境に区切られた領域のうち，下半分の領域内に入った点の数:", N_in)
print("正方形領域に打たれた全ての点の数:", N)

print("実験によって計算される log 2 の近似値:", N_in, "/", N, "=", Log2)
```

##  $\log 2$ の近似値を得る上記のプログラムを 8000 回繰り返し，8000 通りで得られる $\log 2$ の近似値を下記のプログラムによってヒストグラムで図示する（以降，$N=100$ とする）．

```python
%matplotlib inline
import matplotlib.pyplot as plt

Log2_list = []

M = 8000
N = 100 # 試行回数

for i in range(0,M):
    
    N_in = 0
    for i in range(0,N):

        ran_x = random.random() + 1 # X の乱数（1 以上 2 未満）
        ran_y = random.random() # Y の乱数（0 以上 1 未満）

        if ran_y <= 1/ran_x:
            N_in += 1

    Log2 = N_in/N # log 2 の近似式
    Log2_list.append(Log2)

### log 2 の近似値のばらつきに関するヒストグラム ###
plt.hist(Log2_list, rwidth=0.9)
plt.title('Approx. values of log 2',{"fontsize":15})
plt.xlabel('Calculated values')
plt.show() # グラフ表示
```
