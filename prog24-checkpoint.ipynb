## 11.1 „É¢„É≥„ÉÜ„Ç´„É´„É≠Ê≥ï„Å®„ÅØÔºàp.128ÔΩûÔºâ<br><br>„ÄÄ„ÅÇ„Çã‰∫ãË±°„ÅåÁ¢∫ÁéáÁöÑ„Å™ÊåØ„ÇãËàû„ÅÑ„Çí„Åô„Çã„Å®„Åç„Å´Ôºå‰π±Êï∞„ÇÑÁ¢∫ÁéáÂàÜÂ∏É„Çí‰Ωø„Å£„Å¶Ëß£„ÇíÂæó„Çà„ÅÜ„Å®„Åô„Çã„Ç¢„É´„Ç¥„É™„Ç∫„É†„ÅÆÁ∑èÁß∞„Çí„Äå„É¢„É≥„ÉÜ„Ç´„É´„É≠Ê≥ï„Äç„Å®„ÅÑ„ÅÜÔºé‰ªäÂõû„ÅØÔºå„É¢„É≥„ÉÜ„Ç´„É´„É≠Ê≥ï„ÇíÁî®„ÅÑ„Å¶Ê¨°„ÅÆÔºí„Å§„ÅÆÂÆüÈ®ì„ÇíË°å„ÅÜÔºé<br><br>ÂÆüÈ®ìÔºëÔºéÂÜÜÂë®Áéá $\pi=3.1415926535...$ „ÅÆËøë‰ººÂÄ§„ÇíË™ø„Åπ„ÇãÔºé<br><br>‰∏ãÂõ≥„ÅÆ„Çà„ÅÜ„Å´ÔºåÊ®™Ëª∏„ÅÆÂå∫Èñì„Çí $[-1,1]$ „Å®„Åó„Åü‰∏ÄËæ∫„ÅÆÈï∑„Åï„Åå 2 „ÅÆÊ≠£ÊñπÂΩ¢„Å´ÂÜÖÊé•„Åô„ÇãÂçäÂæÑ 1 „ÅÆÂÜÜ„ÅßÂõ≤„Åæ„Çå„ÅüÈ†òÂüü„ÇíËÄÉ„Åà„ÇãÔºé

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np

### ÂÜÜ„ÅÆ‰ΩúÂõ≥ ###
def f1(x):
    return (1-x**2)**(1/2) # ‰∏äÂçäÂàÜ„ÅÆÂÜÜ„ÅÆÊñπÁ®ãÂºè

def f2(x):
    return -(1-x**2)**(1/2) # ‰∏ãÂçäÂàÜ„ÅÆÂÜÜ„ÅÆÊñπÁ®ãÂºè

x = np.linspace(-1, 1, 201)

### „Ç∞„É©„ÉïÊèèÁîª ###
plt.plot(x, f1(x), color = "black")
plt.plot(x, f2(x), color = "black")

plt.axis('square') # Ë°®Á§∫„Åï„Çå„ÇãÂõ≥„ÅÆÁ∏¶Ê®™ÊØî„Çí 1:1 „Å´„Åô„Çã
plt.xlim(-1, 1)
plt.ylim(-1, 1)

plt.show()
```

## Ê¨°„Å´ÔºåÊ≠£ÊñπÂΩ¢„ÅÆ‰∏≠„Å´Ôºå„É©„É≥„ÉÄ„É†„Å´ $N$ ÂÄã„ÅÆÁÇπ„ÇíÊâì„Å§ÔºéÈ†òÂüüÂÖ®‰Ωì„ÅÆÊ≠£ÊñπÂΩ¢ÂÜÖ„Å´Êâì„Åü„Çå„ÅüÁÇπ„ÅåÂÜÜÂÜÖ„Å´„ÅÇ„ÇãÁ¢∫Áéá„ÅØ<br><br>ÂÜÜ„ÅÆÈù¢Á©ç $\div$ Ê≠£ÊñπÂΩ¢„ÅÆÈù¢Á©ç $= \pi \div 4 = \frac{\pi}{4}$ <br><br>„Å®„Å™„ÇãÔºéÂÆüÈöõÔºåÁÇπ„ÅÆÂÄãÊï∞ $N$ „ÇíÂ§ß„Åç„Åè„Å®„Çã„Å®Ôºå<br><br>ÂÜÜÂÜÖ„Å´Êâì„Åü„Çå„ÅüÁÇπ„ÅÆÊï∞ $N_{\rm in}$  $\div$ È†òÂüüÂÖ®‰Ωì„ÅÆÊ≠£ÊñπÂΩ¢ÂÜÖ„Å´Êâì„Åü„Çå„ÅüÂÖ®„Å¶„ÅÆÁÇπ„ÅÆÊï∞ $N$ $\simeq \frac{\pi}{4}$ <br><br>„Åß„ÅÇ„Çã„Åì„Å®„ÅåÁü•„Çâ„Çå„Å¶„ÅÑ„ÇãÔºé„Åì„ÅÆ„Åì„Å®„Åã„ÇâÔºåÊ¨°„ÅÆÂºèÔºö<br><br>$\pi \simeq$ÔºàÂÜÜÂÜÖ„Å´Êâì„Åü„Çå„ÅüÁÇπ„ÅÆÊï∞ $N_{\rm in}$  $\div$ È†òÂüüÂÖ®‰Ωì„ÅÆÊ≠£ÊñπÂΩ¢ÂÜÖ„Å´Êâì„Åü„Çå„ÅüÂÖ®„Å¶„ÅÆÁÇπ„ÅÆÊï∞ $N$Ôºâ$\times$ $4$ <br><br>„ÇíËÄÉ„Åà„Çã„Åì„Å®„ÅßÔºå$\pi$ „ÅÆËøë‰ººÂÄ§„ÇíË™ø„Åπ„Çã„Åì„Å®„ÅåÂèØËÉΩ„Å®„Å™„ÇãÔºé

## „ÄêÁ¢∫Ë™ç„Äë$N=100000$ „ÅÆÂ†¥Âêà

```python
N_in = 0

N = 100000 # Ë©¶Ë°åÂõûÊï∞

for i in range(0,N):

    ran_x = 2 * random.random() - 1 # X „ÅÆ‰π±Êï∞Ôºà-1 ‰ª•‰∏ä 1 Êú™Ê∫ÄÔºâ
    ran_y = 2 * random.random() - 1 # Y „ÅÆ‰π±Êï∞Ôºà-1 ‰ª•‰∏ä 1 Êú™Ê∫ÄÔºâ

    ran_point = (ran_x**2 + ran_y**2) ** (1/2) # ÂéüÁÇπ (0,0) „Åã„Çâ (ran_x,ran_y) „Åæ„Åß„ÅÆË∑ùÈõ¢  

    # ÂéüÁÇπ (0,0) „Åã„Çâ (ran_x,ran_y) „Åæ„Åß„ÅÆË∑ùÈõ¢„Åå 1 ‰ª•‰∏ãÔºå„Åô„Å™„Çè„Å°Ôºå
    # (ran_x,ran_y) „ÅåÂçäÂæÑ 1 „ÅÆÂÜÜÂÜÖ„Å´„ÅÇ„Çå„Å∞ÔºåÂÜÜÈ†òÂüüÂÜÖ„Å´ÂÖ•„Å£„ÅüÁÇπ„ÅÆÊï∞ N_in „Å´ 1 „ÇíÂä†„Åà„Çã
    if ran_point <= 1: 
        N_in += 1

Pie = N_in/N*4 # œÄ„ÅÆËøë‰ººÂºè

print("ÂÜÜÈ†òÂüüÂÜÖ„Å´ÂÖ•„Å£„ÅüÁÇπ„ÅÆÊï∞ N_in=", N_in)
print("Ê≠£ÊñπÂΩ¢È†òÂüü„Å´Êâì„Åü„Çå„ÅüÂÖ®„Å¶„ÅÆÁÇπ„ÅÆÊï∞ N=", N)

print("ÂÆüÈ®ì„Å´„Çà„Å£„Å¶Ë®àÁÆó„Åï„Çå„ÇãœÄ„ÅÆËøë‰ººÂÄ§ N_in / N * 4 =", N_in, "/", N, "* 4 =", N_in/N*4)
```

## $\pi$ „ÅÆËøë‰ººÂÄ§„ÇíÂæó„Çã‰∏äË®ò„ÅÆ„Éó„É≠„Ç∞„É©„É†„Çí 8000 ÂõûÁπ∞„ÇäËøî„ÅóÔºå8000 ÈÄö„Çä„ÅßÂæó„Çâ„Çå„Çã $\pi$ „ÅÆËøë‰ººÂÄ§„Çí‰∏ãË®ò„ÅÆ„Éó„É≠„Ç∞„É©„É†„Å´„Çà„Å£„Å¶„Éí„Çπ„Éà„Ç∞„É©„É†„ÅßÂõ≥Á§∫„Åô„ÇãÔºà‰ª•ÈôçÔºå$N=100$ „Å®„Åô„ÇãÔºâÔºé

```python
%matplotlib inline
import matplotlib.pyplot as plt

Pie_list = []

M = 8000
N = 100 # Ë©¶Ë°åÂõûÊï∞

for i in range(0,M):
    
    N_in = 0
    for j in range(0,N):

        ran_x = 2*random.random()-1 # X „ÅÆ‰π±Êï∞Ôºà-1 ‰ª•‰∏ä 1 Êú™Ê∫ÄÔºâ
        ran_y = 2*random.random()-1 # Y „ÅÆ‰π±Êï∞Ôºà-1 ‰ª•‰∏ä 1 Êú™Ê∫ÄÔºâ

        ran_point = (ran_x**2 + ran_y**2) ** (1/2) # ÂéüÁÇπ (0,0) „Åã„Çâ (ran_x,ran_y) „Åæ„Åß„ÅÆË∑ùÈõ¢  

        if ran_point <= 1:
            N_in += 1

    Pie = N_in/N*4 # œÄ„ÅÆËøë‰ººÂºè
    Pie_list.append(Pie)

### ùúã „ÅÆËøë‰ººÂÄ§„ÅÆ„Å∞„Çâ„Å§„Åç„Å´Èñ¢„Åô„Çã„Éí„Çπ„Éà„Ç∞„É©„É† ###
plt.hist(Pie_list, rwidth=0.9)
plt.title('Approx. values of $\pi$',{"fontsize":15})
plt.xlabel('Calculated values')
plt.show() # „Ç∞„É©„ÉïË°®Á§∫
```

## ÂÆüÈ®ìÔºíÔºé$\log 2 (=\log_e 2)=0.69314718...$ „ÅÆËøë‰ººÂÄ§„ÇíË™ø„Åπ„ÇãÔºé<br><br>‰∏ãÂõ≥„ÅÆ„Çà„ÅÜ„Å´ÔºåÊ®™Ëª∏„ÅÆÂå∫Èñì„Çí $[1,2]$ „Å®„Åó„Åü‰∏ÄËæ∫„ÅÆÈï∑„Åï„Åå 1 „ÅÆÊ≠£ÊñπÂΩ¢„ÅÆ„ÅÜ„Å°ÔºåÊõ≤Á∑ö $y=\frac{1}{x}$ „ÇíÂ¢É„Å´Âå∫Âàá„Çâ„Çå„ÅüÈ†òÂüü„ÅÆ„ÅÜ„Å°Ôºå‰∏ãÂçäÂàÜ„ÅÆÈ†òÂüü„ÇíËÄÉ„Åà„ÇãÔºé

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np

### Êõ≤Á∑ö y=1/x „ÅÆ‰ΩúÂõ≥ ###
def f(x):
    return 1/x

x = np.linspace(1, 2, 201)

### „Ç∞„É©„ÉïÊèèÁîª ###
plt.plot(x, f(x), color = "black")

plt.axis('square') # Ë°®Á§∫„Åï„Çå„ÇãÂõ≥„ÅÆÁ∏¶Ê®™ÊØî„Çí 1:1 „Å´„Åô„Çã
plt.xlim(1, 2)
plt.ylim(0, 1)

plt.show()
```

## Ê≥®ÊÑèÔºéÊõ≤Á∑ö $y=\frac{1}{x}$ „ÇíÂ¢É„Å´Âå∫Âàá„Çâ„Çå„ÅüÈ†òÂüü„ÅÆ„ÅÜ„Å°Ôºå‰∏ãÂçäÂàÜ„ÅÆÈ†òÂüü„ÅÆÈù¢Á©ç„Çí $S$ „Å®„Åä„Åè„Å®Ôºå$$S=\int_1^2 \frac{1}{x} \ dx=[ \log x ]_1^2=\log 2-0=\log 2 \ (=\log_e 2)$$ „Çà„ÇäÔºå $S=\log 2$ „Åß„ÅÇ„ÇãÔºé<br><br>„ÄÄÊ¨°„Å´ÔºåÊ≠£ÊñπÂΩ¢„ÅÆ‰∏≠„Å´Ôºå„É©„É≥„ÉÄ„É†„Å´ $N$ ÂÄã„ÅÆÁÇπ„ÇíÊâì„Å§ÔºéÈ†òÂüüÂÖ®‰Ωì„ÅÆÊ≠£ÊñπÂΩ¢ÂÜÖ„Å´Êâì„Åü„Çå„ÅüÁÇπ„ÅåÔºåÊõ≤Á∑ö $y=\frac{1}{x}$ „ÇíÂ¢É„Å´Âå∫Âàá„Çâ„Çå„ÅüÈ†òÂüü„ÅÆ„ÅÜ„Å°Ôºå‰∏ãÂçäÂàÜ„ÅÆÈ†òÂüüÂÜÖ„Å´„ÅÇ„ÇãÁ¢∫Áéá„ÅØ<br><br>$S$ $\div$ Ê≠£ÊñπÂΩ¢„ÅÆÈù¢Á©ç $= \log 2 \div 1 = \log 2$ <br><br>„Å®„Å™„ÇãÔºéÂÆüÈöõÔºåÁÇπ„ÅÆÂÄãÊï∞ $N$ „ÇíÂ§ß„Åç„Åè„Å®„Çã„Å®Ôºå<br><br>Êõ≤Á∑ö $y=\frac{1}{x}$ „ÇíÂ¢É„Å´Âå∫Âàá„Çâ„Çå„ÅüÈ†òÂüü„ÅÆ„ÅÜ„Å°Ôºå‰∏ãÂçäÂàÜ„ÅÆÈ†òÂüüÂÜÖ„Å´Êâì„Åü„Çå„ÅüÁÇπ„ÅÆÊï∞ $N_{\rm in}$<br>„ÄÄ$\div$ È†òÂüüÂÖ®‰Ωì„ÅÆÊ≠£ÊñπÂΩ¢ÂÜÖ„Å´Êâì„Åü„Çå„ÅüÂÖ®„Å¶„ÅÆÁÇπ„ÅÆÊï∞ $N$ $\simeq \log 2$ <br><br>„Åß„ÅÇ„Çã„Åì„Å®„ÅåÁü•„Çâ„Çå„Å¶„ÅÑ„ÇãÔºé„Åì„ÅÆ„Åì„Å®„Åã„ÇâÔºåÊ¨°„ÅÆÂºèÔºö<br><br>$\log 2 \simeq$ Êõ≤Á∑ö $y=\frac{1}{x}$ „ÇíÂ¢É„Å´Âå∫Âàá„Çâ„Çå„ÅüÈ†òÂüü„ÅÆ„ÅÜ„Å°Ôºå‰∏ãÂçäÂàÜ„ÅÆÈ†òÂüüÂÜÖ„Å´Êâì„Åü„Çå„ÅüÁÇπ„ÅÆÊï∞ $N_{\rm in}$<br>„ÄÄ„ÄÄ„ÄÄ„ÄÄ$\div$ È†òÂüüÂÖ®‰Ωì„ÅÆÊ≠£ÊñπÂΩ¢ÂÜÖ„Å´Êâì„Åü„Çå„ÅüÂÖ®„Å¶„ÅÆÁÇπ„ÅÆÊï∞ $N$<br><br>„ÇíËÄÉ„Åà„Çã„Åì„Å®„ÅßÔºå$\log 2$ „ÅÆËøë‰ººÂÄ§„ÇíË™ø„Åπ„Çã„Åì„Å®„ÅåÂèØËÉΩ„Å®„Å™„ÇãÔºé

## „ÄêÁ¢∫Ë™ç„Äë$N=100000$ „ÅÆÂ†¥Âêà

```python
N = 100000 # Ë©¶Ë°åÂõûÊï∞

N_in = 0

for i in range(0,N):

    ran_x = random.random() + 1 # X „ÅÆ‰π±Êï∞Ôºà1 ‰ª•‰∏ä 2 Êú™Ê∫ÄÔºâ
    ran_y = random.random() # Y „ÅÆ‰π±Êï∞Ôºà0 ‰ª•‰∏ä 1 Êú™Ê∫ÄÔºâ

    if ran_y <= 1/ran_x:
        N_in += 1

Log2 = N_in/N # log 2 „ÅÆËøë‰ººÂºè

print("Êõ≤Á∑ö y=1/x „ÇíÂ¢É„Å´Âå∫Âàá„Çâ„Çå„ÅüÈ†òÂüü„ÅÆ„ÅÜ„Å°Ôºå‰∏ãÂçäÂàÜ„ÅÆÈ†òÂüüÂÜÖ„Å´ÂÖ•„Å£„ÅüÁÇπ„ÅÆÊï∞:", N_in)
print("Ê≠£ÊñπÂΩ¢È†òÂüü„Å´Êâì„Åü„Çå„ÅüÂÖ®„Å¶„ÅÆÁÇπ„ÅÆÊï∞:", N)

print("ÂÆüÈ®ì„Å´„Çà„Å£„Å¶Ë®àÁÆó„Åï„Çå„Çã log 2 „ÅÆËøë‰ººÂÄ§:", N_in, "/", N, "=", Log2)
```

##  $\log 2$ „ÅÆËøë‰ººÂÄ§„ÇíÂæó„Çã‰∏äË®ò„ÅÆ„Éó„É≠„Ç∞„É©„É†„Çí 8000 ÂõûÁπ∞„ÇäËøî„ÅóÔºå8000 ÈÄö„Çä„ÅßÂæó„Çâ„Çå„Çã $\log 2$ „ÅÆËøë‰ººÂÄ§„Çí‰∏ãË®ò„ÅÆ„Éó„É≠„Ç∞„É©„É†„Å´„Çà„Å£„Å¶„Éí„Çπ„Éà„Ç∞„É©„É†„ÅßÂõ≥Á§∫„Åô„ÇãÔºà‰ª•ÈôçÔºå$N=100$ „Å®„Åô„ÇãÔºâÔºé
