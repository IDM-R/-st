## 7.3 計算結果をもとにグラフを描く（p.91～）<br><br>7.3.1 関数のグラフを描く<br><br>グラフの横軸に対応する $x$ を -10 から 10 まで 1 ずつ変化させて，各 $x$ の値に対応する $x^2-1$ の値を計算し，折れ線グラフにする．その準備として，次の Code セルを実行することで，横軸の値のリストを始めに作成する．

```python
x=list(range(-10,11))
```

## 続いて，$x^2-1$ を求める関数 f() を定義して，上記のリスト x に map() を適用する．ここで，map() は，リストの各要素に与えられた関数を適用し，その結果を集めたオブジェクトを返す．list() 関数と併用した次の Code セルを実行することで，計算結果のリストを作成できる．

```python
#講義（前期，第１３回）レポート，問３や例２も参照せよ．

def f(x):
    return x**2-1
y = list(map(f,x))
y
```

## 作成したリスト $x$ と $y$ の各要素の値の組を，plot() を使ってプロットする.　

```python
plt.plot(x,y)
```

## ＜軸や凡例などに関するオプション指定＞<br><br>（１）横軸の下に独立変数 $x$，縦軸の左に関数を添えたい場合

```python
plt.plot(x,y)

#### オプション ###
plt.xlabel('$x$') # 軸の下側に表示される横軸の説明
plt.ylabel('$f(x)$') # 軸の左側に表示される縦軸の説明
```

## （２）横軸の下に独立変数 $x$，縦軸の左に従属変数 $y$ を添えて，グラフの右上に説明を加えたい場合<br><br>・グラフが１つである場合

```python
plt.plot(x,y, label="$y=f(x)$") # 凡例に用いる曲線 y=f(x) の名前を label オプションで指定

#### オプション ###
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # 図全体の左下の座標を(0,0), 右上の座標を(1,1)とした凡例（legend）の位置調整
```

## ・グラフが２つ以上である場合（$f_1(x)=x-20,f_2(x)=100-3x^2$）

```python
x=list(range(-10,11))

#### 関数定義 ###
def f1(x):
    return x-20
y1 = list(map(f1,x))
y1

def f2(x):
    return 100-3*x**2
y2 = list(map(f2,x))
y2

### グラフ描画 ###
plt.plot(x,y1, label="$y=f_1(x)$") # 凡例に用いる曲線 y=f1(x) の名前を label オプションで指定
plt.plot(x,y2, label="$y=f_2(x)$") # 凡例に用いる曲線 y=f2(x) の名前を label オプションで指定

#### オプション ###
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # 図全体の左下の座標を(0,0), 右上の座標を(1,1)とした凡例（legend）の位置調整
```

## 【特別編１】numpy パッケージの linspace() を使った，滑らかなグラフの描画<br><br>次のグラフは，$x$ を -2 から 2 まで 1 ずつ変化させて，各 $x$ の値に対応する $x^2-1$ の値を計算し，折れ線グラフにしたものである．

```python
x1=list(range(-2,3,1))

def f(x):
    return x**2-1
y = list(map(f,x1))

plt.plot(x1,y)
```

## 上のグラフの場合，$x$ を -2 から 2 まで 1 ずつ変化させるために，range() の第３引数が 1 となっている．しかし，この場合にプロットされる点は (-2,4), (-1,1), (0,0), (1,1), (2,4) の５点しかない上，range() の第３引数は整数値しかとれない．そのため，グラフが滑らかにならない．一方，基本的な数値計算機能を提供する numpy と呼ばれるパッケージの linspace() を使うことで，プロットしたい点の $x$ 座標の間隔を<span style="color: blue; ">非整数値</span>で指定できる．

```python
import numpy as np # numpy をインポート
np.linspace(-2, 2, 4) # 第１引数：x 座標の始点，第２引数：x 座標の終点，第３引数：始点から終点までの選点の個数
```

```python
import numpy as np
np.linspace(-2, 2, 10)
```

```python
import numpy as np
np.linspace(-2, 2, 201)
```

## linspace() を使えば，区間の幅によることなく，プロットしたい点の $x$ 座標の間隔を短くとることで，滑らかなグラフを表示させることができる．<br><br>例１－１．linspace() の第３引数を 4 としたグラフ<br>（プロットされる点の $x$ 座標の間隔 $=\frac{2-(-2)}{4-1}=$<span style="color: blue; ">1.333...</span>）

```python
import numpy as np

### 関数定義 ###
def f(x):
    return x**2-1

### 横軸の区間と選点数の設定 ###
x = np.linspace(-2, 2, 4)

### グラフ描画 ###
plt.plot(x, f(x))
```

## 例１－２．linspace() の第３引数を 10 としたグラフ<br>（プロットされる点の $x$ 座標の間隔 $=\frac{2-(-2)}{10-1}=$<span style="color: blue; ">0.444...</span>）

```python
import numpy as np

### 関数定義 ###
def f(x):
    return x**2-1

### 横軸の区間と選点数の設定 ###
x = np.linspace(-2, 2, 10)

### グラフ描画 ###
plt.plot(x, f(x))
```

## 例１－３．linspace() の第３引数を 201 としたグラフ<br>（プロットされる点の $x$ 座標の間隔 $=\frac{2-(-2)}{201-1}=$<span style="color: blue; ">0.02</span>）

```python
import numpy as np

### 関数定義 ###
def f(x):
    return x**2-1

### 横軸の区間と選点数の設定 ###
x = np.linspace(-2, 2, 201)

### グラフ描画 ###
plt.plot(x, f(x))
```

## 例１－４．linspace() を使った，同一区間における２つの滑らかなグラフ

```python
import numpy as np

### 関数定義 ###
def f1(x):
    return x**2-1
def f2(x):
    return -2*x**3+5

### 横軸の区間と選点数の設定 ###
x = np.linspace(-2, 2, 201)

### グラフ描画 ###
plt.plot(x, f1(x), label="$y=x^2-1$")
plt.plot(x, f2(x), label="$y=-2x^3+5$")

### オプション ###
plt.xlabel('$x$')
plt.ylabel('$y$')
    
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # 図全体の左下の座標を(0,0), 右上の座標を(1,1)とした凡例（legend）の位置調整
```

## 例４．乱数を 100 回生成させる関数 rnd(x) を定義し，折れ線グラフで乱数たちをプロットするプログラム

```python
def rnd(x):
    return random.random()
sample = list(map(rnd, range(100))) # 乱数を100個含むリストを生成

plt.plot(sample)
```
