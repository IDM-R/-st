## 数値計算を用いて，与えられた関数の極大値・極小値を考えよう．今回は，極小値のみ考える．<br><br>　講義（後期，第２回）に倣って，$f(x)=x^2-3x$ として，$y=f(x)$ のグラフを下記の通りに図示する．ここで，grid() はプロット上に目盛りを描画するための命令文である．

```python
import numpy as np
import matplotlib.pyplot as plt

### 関数定義 ###
def f(x):
    return x**2-3*x

### 横軸の区間と選点数の設定 ###
x = np.linspace(-2.5, 5.5, 201)

### グラフ描画 ###
plt.plot(x, f(x), label="$y=x^2-3x$")

### f が極小値をとるときの x の値と f(x) の値の組のプロット ###
plt.scatter(3/2, f(3/2), color="black")

### オプション（ラベル，目盛り） ###
plt.xlabel("$x$") # x 軸の名前
plt.ylabel("$y=f(x)$") # y 軸の名前
plt.grid() # 目盛り表示

### オプション（凡例） ###
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # 図全体の左下の座標を(0,0), 右上の座標を(1,1)とした凡例（legend）の位置調整
```

## $f$ が極小となるような $x$ を $x_{\rm min}$ とする．上のようなシンプルな関数であれば，その導関数を調べることで，$x_{\rm min}=\frac{3}{2}$ で $f$ は極小値をとり，$f(x_{\rm min})=-\frac{9}{4}$ となることが容易にわかる．しかし，$f$ がより一般の関数となると，$f$ の極小値や $x_{\rm min}$ を具体的に求めることは難しい．こうした場合は，数値計算による探索が有効である．<br>　$f(x)$ を小さくするために $x$ を動かしながら効率よく $x_{\rm min}$ を見つけるには，下記の２点：<br><br>（１）$x$ を正負どちらの向きに動かせば良いか？<br>（２）$x$ をどれだけ大きく動かせばよいか？<br><br>を決めることが重要である．<br><br>＜極小値の探索<span style="color: blue; ">（勾配降下法）</span>＞<br>　$x=-2,-1,...,5$ の各点での導関数 $f^{\prime}(x)$ の値に負の一定数 $-0.2$ をかけた数を（$x$ 軸の正の向きでの）横向きの移動量とする矢印を，グラフ上の各点を始点としてプロットする．

```python
import numpy as np
import matplotlib.pyplot as plt

### 関数定義 ###
def f(x):
    return x**2-3*x

### 横軸の区間と選点数の設定 ###
x = np.linspace(-2.5, 5.5, 201)

### グラフ描画 ###
plt.plot(x, f(x), label="$y=x^2-3x$")

### x=-2,-1,...,5 の各点での微分値 f'(x) に負の定数（-0.3）をかけた数を横向きの移動方向とする矢印のプロット ###
a=0.2
h=1e-10
for x in range(-2,6):
    plt.scatter(x, f(x), color="black")
    Df = (f(x+h)-f(x))/h
    plt.quiver(x,f(x), -a*Df, 0, angles='xy', scale_units='xy', scale=1) # 矢印プロット

### オプション（ラベル，目盛り） ###
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.grid()
plt.scatter(3/2, f(3/2), color="blue", s=100) # s: size

### オプション（凡例） ###
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

## $f$ が極小となるような $x$ を $x_{\rm min}$ とする．上図の場合は $x_{\rm min}=\frac{3}{2}$ であり，次のことがわかる：<br><br>・$x$ が $x_{\rm min}$ から離れている場合，$f^{\prime}(x)$ の値は大きくなり，横向き移動量も大きくなる．<br>・$x$ が $x_{\rm min}$ に近い場合，$f^{\prime}(x)$ の値は小さくなり，横向き移動量も小さくなる．<br><br>$x_{\rm min}$ を探すために与える $x$ の初期点を $x_0$ とおく．<span style="color: red; ">微分 $f^{\prime}(x_0)$ の値に負の定数をかけた値</span>を（$x$ 軸の正の向きでの）$x_0$ からの探索移動量とする．上の２つの性質から，この移動量と探索方向は適度なものとなる．すなわち，各 $n \geq 1$ に対して，$\alpha >0$ を定数として<span style="color: red; ">微分 $f^{\prime}(x_{n-1})$ の値に $-\alpha$ をかけた値</span>を使って $x_{n-1}$ を順次更新する計算：\begin{align*} x_1=x_0&\color{red}{-}\color{red}{\alpha f^{\prime}(x_0)} \\ x_2=x_1&\color{red}{-}\color{red}{\alpha f^{\prime}(x_1)}  \\ \vdots \end{align*} <br>を繰り返す次の再帰関数（講義（前期，第１３回）参照）：<br><br>$$x_n=x_{n-1}\color{red}{-}\color{red}{\alpha f^{\prime}(x_{n-1})}$$<br>から求まる $x_n$ の動きを調べると，$\alpha$ が小さいとき，$x_n \to x_{\rm min}$ $(n \to \infty)$ となる．このアルゴリズムで $x_{\rm min}$ を探索する方法を，<span style="color: blue; ">勾配降下法</span>という．また，パラメータ $\alpha>0$ は，機械学習という分野では「学習率」とよばれている．<br><br>【注意】学習率 $\alpha$ を極端に大きく・小さくとると，次のメリットやデメリットが生じる．<br><br>$\alpha$ が極端に大きいとき：<br>　少ない計算回数で $x_n$ は $x_{\rm min}$ に収束することがある．しかし，$x_n$ が符号変化しながらとびとびで変化し，いつまでも収束しないこともある．<br><br>$\alpha$ が極端に小さいとき：<br>　$x_n$ は $x_{\rm min}$ に収束しやすくなる．しかし，$x_{\rm min}$ に近い値を $x_n$ がとるための計算回数が多くなるために計算コストがかかり，処理時間が極端に長くなることもある．

## 以降の数値計算では，この学習率 $\alpha$ を小さめにとるよう留意している．<br><br>例１－１．2 次関数 $f(x)=x^2-3x$ の極小値の探索

```python
import numpy as np
import matplotlib.pyplot as plt

### 関数定義 ###
def f(x):
    return x**2-3*x

### 勾配降下法の実装 ###
a = 0.2 # 学習率

x0 = -2 # x の初期値
plt.scatter(x0, f(x0), color="green", s=100)

h = 1e-10 # 導関数 f' を与えるための微小な値
error = 1e-8 # 収束判定用の数値
max_loop = 50 # 計算回数の最大値

num_calc = 0 # 計算回数の初期値を 0 とする
print("%3d" % num_calc,"回目の探索：x=", x0)

while(True):
    Df = (f(x0+h)-f(x0))/h
    x1 = x0 - a*Df # 根の近似値の更新
    num_calc += 1 # 計算回数のカウントアップ
    print("%3d" % num_calc, "回目の探索：x=", x1)
    plt.plot(x1, f(x1), marker='.', markersize=10, color="black") # 根の近似値とそのときの f の値を点で表示

    
    # 終了条件１：更新前後で x の増分（の絶対値）が十分小さくなったときに，探索を終了する
    if(abs(x1-x0) <= error):
        print()
        print("x_min の探索移動量：small enough (<=", error, ").")
        plt.plot(x1, f(x1), marker='.', markersize=20, color="blue") # 根の近似値とそのときの f の値を点で表示
        print("x_min=", x1, "，f(x_min)=", f(x1))
        break
        
    # 終了条件２：max_loop 回分の計算を終えた場合でも，探索を終了する
    if(max_loop<=num_calc):
        print("The number of loops reach a maximum value.")
        break
    
    x0 = x1 # 根の近似値を更新

### グラフ描画 ###
x = np.linspace(-2.5, 5.5, 201)
plt.plot(x, f(x))

### オプション ###
plt.xlabel("$x$") # x 軸の名前
plt.ylabel("$f(x)$") # y 軸の名前
plt.grid() # 点線の目盛り表示
```

## 上の例では，<span style="color: green; ">$x_0=-2$</span>, $\alpha=0.2$ とすると，$$ x_1=-0.599..., \ x_2=0.240..., \ \cdots, \ x_{31}=1.499...$$といったように，$n$ を大きくすると $x_n$ は <span style="color: blue; ">$x_{\rm min}=\frac{3}{2}$</span> に近づく．次に，極小値の探索として $x_{\rm min}$ を手計算では容易に求められなかったり，$x_{\rm min}$ が複数ある場合を考えよう．<br><br>例１－２－１．<span style="color: green; ">$x_0=1.7$</span> とした $g(x)=2^{-5x}+\frac{x^2}{10}-\frac{x^3}{25}+\frac{x^5}{1000}$ の極小値の探索（$x_{\rm min}=\color{blue}{1.077...}, 3.582...$）

```python
import numpy as np
import matplotlib.pyplot as plt

### 関数定義 ###
def g(x):
    return 2**(-5*x)+x**2/10-x**3/25+x**5/1000

### 勾配降下法の実装 ###
a = 2 # 学習率

x0 = 1.7 # x の初期値
plt.scatter(x0, g(x0), color="green", s=100)

h = 1e-10 # 導関数 f' を与えるための微小な値
error = 1e-8 # 収束判定用の数値
max_loop = 200 # 計算回数の最大値

num_calc = 0 # 計算回数の初期値を 0 とする
print("%3d" % num_calc, "回目の探索：x=", x0)

while(True):
    Dg = (g(x0+h)-g(x0))/h
    x1 = x0 - a*Dg # 根の近似値の更新
    num_calc += 1 # 計算回数のカウントアップ
    print("%3d" % num_calc,"回目の探索：x=", x1)
    plt.plot(x1, g(x1), marker='.', markersize=10, color="black") # 根の近似値とそのときの f の値を点で表示

    # 終了条件１：更新前後で x の増分（の絶対値）が十分小さくなったときに，探索を終了する
    if(abs(x1-x0) <= error):
        print()
        print("x_min の探索移動量：small enough (<=", error, ").")
        plt.plot(x1, g(x1), marker='.', markersize=20, color="blue") # 根の近似値とそのときの f の値を点で表示
        print("x_min=", x1, "，g(x_min)=", g(x1))
        break
        
    # 終了条件２：max_loop 回分の計算を終えた場合でも，探索を終了する
    if(max_loop<=num_calc):
        print("The number of loops reach a maximum value.")
        break
    
    x0 = x1 # 根の近似値を更新

### グラフ描画 ###
x = np.linspace(0.5, 4.5, 201)
plt.plot(x, g(x))

### オプション ###
plt.xlabel("$x$") # x 軸の名前
plt.ylabel("$g(x)$") # y 軸の名前
plt.grid() # 点線の目盛り表示
```

## 上の例では，<span style="color: green; ">$x_0=1.7$</span> , $\alpha=2$ とすると，<br> $$ x_1=1.649..., \ x_2=1.591..., \ \cdots, \ x_{26}=1.077...$$といったように，$n$ を大きくすると $x_n$ は <span style="color: blue; ">$x_{\rm min}$ のうちの１つの値 $1.077...$</span> に近づく．<br><br>例１－２－２．<span style="color: green; ">$x_0=2.2$</span> とした $g(x)=2^{-5x}+\frac{x^2}{10}-\frac{x^3}{25}+\frac{x^5}{1000}$ の極小値の探索（$x_{\rm min}=1.077..., \color{blue}{3.582...}$）

```python
import numpy as np
import matplotlib.pyplot as plt

### 関数定義 ###
def g(x):
    return 2**(-5*x)+x**2/10-x**3/25+x**5/1000

### 勾配降下法の実装 ###
a = 2 # 学習率

x0 = 2.2 # x の初期値
plt.scatter(x0, g(x0), color="green", s=100)

h = 1e-10 # 導関数 f' を与えるための微小な値
error = 1e-8 # 収束判定用の数値
max_loop = 200 # 計算回数の最大値

num_calc = 0 # 計算回数の初期値を 0 とする
print("%3d" % num_calc, "回目の探索：x=", x0)

while(True):
    Dg = (g(x0+h)-g(x0))/h
    x1 = x0 - a*Dg # 根の近似値の更新
    num_calc += 1 # 計算回数のカウントアップ
    print("%3d" % num_calc,"回目の探索：x=", x1)
    plt.plot(x1, g(x1), marker='.', markersize=10, color="black") # 根の近似値とそのときの f の値を点で表示

    # 終了条件１：更新前後で x の増分（の絶対値）が十分小さくなったときに，探索を終了する
    if(abs(x1-x0) <= error):
        print()
        print("x_min の探索移動量：small enough (<=", error, ").")
        plt.plot(x1, g(x1), marker='.', markersize=20, color="blue") # 根の近似値とそのときの f の値を点で表示
        print("x_min=", x1, "，g(x_min)=", g(x1))
        break
        
    # 終了条件２：max_loop 回分の計算を終えた場合でも，探索を終了する
    if(max_loop<=num_calc):
        print("The number of loops reach a maximum value.")
        break
    
    x0 = x1 # 根の近似値を更新

### グラフ描画 ###
x = np.linspace(0.5, 4.5, 201)
plt.plot(x, g(x))

### オプション ###
plt.xlabel("$x$") # x 軸の名前
plt.ylabel("$g(x)$") # y 軸の名前
plt.grid() # 点線の目盛り表示
```

## 上の例では，<span style="color: green; ">$x_0=2.2$</span> , $\alpha=2$ とすると，<br> $$ x_1=2.250..., \ x_2=2.312..., \ \cdots, \ x_{31}=3.582...$$といったように，$n$ を大きくすると $x_n$ は <span style="color: blue; ">$x_{\rm min}$ のうちの１つの値 $3.582...$</span> に近づく．勾配降下法で見つかる $x_{\rm min}$ は，グラフ上の適当な所にボールを置いて手を離し，斜面に沿って転げ降りながら，<span style="color: blue; ">最終的にボールの動きが落ち着く所の $x$ 座標</span>と考えて良い．
