## に基づいて，0 ステップ目から <span style="color: blue; ">50 ステップ</span>目までの位置の移動経過を，講義ノート最下部にならって，<span style="color: blue; ">アニメーションにより</span>図示せよ．必要に応じて，複数の Code セルを用いて良い．

```python
# 問１の入力欄
def next_step():
    if 0<= random.random()  < 0.25:
        return 1,0
    elif random.random() < 0.50:
        return 0,1
    elif random.random() < 0.75:
        return -1,0
    else:
        return 0,-1
```

```python
%matplotlib notebook

import matplotlib.pyplot as plt
import matplotlib.animation as animation 

fig = plt.figure()
ims = []

posx = 0
posy = 0
posx_history = [0] 
posy_history = [0] 

for i in range(50):
    dx, dy = next_step()
    posx += dx
    posy += dy
    posx_history.append(posx) 
    posy_history.append(posy) 
    im = plt.plot(posx_history, posy_history, marker='o', color="black", markersize=5) 
    ims.append(im)
    
plt.plot(0, 0, marker='.', markersize=20) 
plt.plot(posx_history[-1], posy_history[-1], marker='s', markersize=10) 

plt.title('Position for each step',{"fontsize":15})
plt.xlabel('$x$-axis')
plt.ylabel('$y$-axis')

ani = animation.ArtistAnimation(fig, ims, interval=100)

plt.show() 
```
