```python
from google_play_scraper import app
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime as dt
```

```python
from google_play_scraper import Sort,reviews_all
```

„É¨„Éì„É•„Éº„Çí„Çπ„ÇØ„É¨„Ç§„Éî„É≥„Ç∞„Åô„Çã„ÄÇ

```python
threads_reviews=reviews_all(
    'com.instagram.barcelona',
    sleep_milliseconds=0,  
    lang='en',  
    country='us',  
    sort=Sort.NEWEST,  
)
```

```python
threads_reviews
```

```python
df_threads = pd.DataFrame(np.array(threads_reviews), columns=['review'])
df_threads = df_threads.join(pd.DataFrame(df_threads.pop('review').tolist()))
df_threads
```

```python
df_t = df_threads[['at', 'score', 'content']]
```

```python
df_t1 = df_t[['at','content']]
df_t1 = df_t[(df_t['at'] > dt.datetime(2023,12,1)) & (df_t1['at'] < dt.datetime(2024,1,14))]
df_t1
```

„É¨„Éì„É•„Éº„Éá„Éº„Çø„Åã„Çâ„ÉÜ„Ç≠„Çπ„Éà„Éá„Éº„Çø„ÇíÂèñ„ÇäÂá∫„Åô„ÄÇ

```python
df_t1 = df_t1[['content']]
df_t1 = df_t1.rename(columns={'content': 'text'})
df_t1
```

Âèñ„ÇäÂá∫„Åó„Åü„É¨„Éì„É•„Éº„ÅÆ„ÉÜ„Ç≠„Çπ„Éà„Éá„Éº„Çø„ÇíÁµêÂêà„Åô„Çã„ÄÇ

```python
import pandas as pd

def combine_text_column(data_frame, column_name):
    combined_text = " ".join(data_frame[column_name].astype(str))
    return combined_text

combined_text_result_t = combine_text_column(df_t1, 'text')

print(combined_text_result_t)
```

```python
t_sp=combined_text_result_t.split()
t_sp[10]
```

ÂçòË™û„ÇíÊï∞„Åà‰∏ä„Åí„Çã„ÄÇ

```python
import collections

d_t = collections.Counter(t_sp)
print(d_t)
```

```python
c_t=d_t.most_common(1000)
```

ÊúâÊÑèÁæ©„Å™ÂçòË™û„ÇíÂê´„ÇÄ„Çà„ÅÜ„Å™ÁØÑÂõ≤„Åß„Ç∞„É©„Éï„Åù‰ΩúÊàê

```python
x = []
y = []

for i in range(30,60):
    x.append(c_t[i][0]) # Ê®™Ëª∏„ÅÆ x Â∫ßÊ®ô„Å®„Åó„Å¶ÔºåÂçòË™û„ÅÆÂá∫ÁèæÂõûÊï∞„ÇíË°®„Åô„É™„Çπ„Éà„Çí‰ΩúÊàê„Åô„Çã
    y.append(c_t[i][1]) # Á∏¶Ëª∏„ÅÆ y Â∫ßÊ®ô„Å®„Åó„Å¶ÔºåÂçòË™û„Åù„ÅÆ„ÇÇ„ÅÆ„ÇíË°®„Åô„É™„Çπ„Éà„Çí‰ΩúÊàê„Åô„Çã

plt.barh(x, y, height = 0.7)
```
