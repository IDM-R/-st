## 7.2.4 円グラフ（Pie Chart）<br><br>例４－１．アンケート調査の結果

```python
plt.axis('equal') # x 軸と y 軸の比率を１：１にして，グラフ全体の形を円にする
data = [15,30,50] # リスト内の数値を足して 100 にならなくても良い
label = ["Yes","No","Unknown"]
plt.pie(data, labels=label)
```

## 例４－２．日本の 2022 年度補正後予算（tn：trillion，兆［円］）<br><font size="4">＜出典＞財務省，日本の財政構造，URL：https://www.mof.go.jp/zaisei/current-situation/index.html </font>

```python
plt.axis('equal')
data = [36.3,15.9,6.1,5.4,5.4,17.0,24.3] # 成分の単位は "兆［円］"
label = ["Social security (36.3 tn)","Local allocation tax, local grant, etc. (15.9 tn)","Public works (6.1 tn)","Expenditure on education and science (5.4 tn)","Defence (5.4 tn)","Others (17.0 tn)","Debt servicing costs (24.3 tn)"]
plt.pie(data, labels=label)
```

## ＜各要素の色，構成割合の表示に関するオプション指定＞<br><br>（１）各要素の色を変えたい場合<br><font size="4">例４．１のプログラムの最終行 "plt.pie(data, labels=label)" を "plt.pie(data, labels=label<span style="color: blue; ">, colors=color</span>)" として，１つ手前の行に，色指定のオプションとして，"<span style="color: blue; ">color = ["lightpink", "magenta", "gold"]</span>" という文を追記している．</font>

```python
plt.axis('equal')
data = [15,30,50]
label = ["Yes","No","Unknown"]
color = ["lightpink", "magenta", "gold"] # 色指定
plt.pie(data, labels=label, colors=color)
```

## （２）構成割合を表示したい場合<br><font size="4">例４．２のプログラムの最終行 "plt.pie(data, labels=label)" を "plt.pie(data, labels=label, <span style="color: blue; ">autopct="%1.3f%%"</span>)" として，構成割合を小数点以下 3 桁まで表示させている（上で提示した財務省の URL にアクセスし，構成割合が正しいかどうかを確認してみましょう）．</font>

```python
plt.axis('equal')
data = [36.3,15.9,6.1,5.4,5.4,17.0,24.3]
label = ["Social security (36.3 tn)","Local allocation tax, local grant, etc. (15.9 tn)","Public works (6.1 tn)","Expenditure on education and science (5.4 tn)","Defence (5.4 tn)","Others (17.0 tn)","Debt servicing costs (24.3 tn)"]
plt.pie(data, labels=label, autopct="%1.3f%%") # 構成割合を小数点以下 3 桁まで表示
```
