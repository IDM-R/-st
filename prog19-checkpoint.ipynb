## はじめに，ソート対象のリスト list1 を定義する．

```python
list1 = [3, 6, 1, 9, 2] # ソート対象
```

## 以下で紹介する２通りの解法によって，リスト内の数を小さい順に並べ替える

## 【解法１（非実用的）】<br><br>Step 1．インデックス s 以降で，リスト l（エル）の要素の最小値があるインデックスを得る関数 min_idx(l, s) を定義する．

```python
def min_idx(l, s):

    # 仮にインデックス s の要素を最小値として，
    # その値を m とすると共に，インデックス s を m_idx とする
    m = l[s]  
    m_idx = s
    
    # i を大きくしながら，より小さい l[i] の値が見つかるごとに，
    # その l[i] を m に上書きすると共にインデックス i を m_idx に上書きする
    for i in range(s, len(l)):
        if m > l[i]: 
            m = l[i]
            m_idx = i
    
    # 最終的に得られる要素の最小値があるインデックスを出力する
    return m_idx
```

```python
# インデックス 0 以降のリスト [3,6,1,9,2] の要素 3,6,1,9,2 の最小値があるインデックス番号を出力する
min_idx(list1, 0)
```

```python
# インデックス 2 以降のリスト [3,6,1,9,2] の要素 1,9,2 の最小値があるインデックス番号を出力する
min_idx(list1, 2)
```

```python
# インデックス 3 以降のリスト [3,6,1,9,2] の要素 9,2 の最小値があるインデックス番号を出力する
min_idx(list1, 3)
```

## Step 2-1．リスト [3, 6, 1, 9, 2] において，<br><br>・インデックス 0 の要素は 3 である．<br>・インデックス 0 以降で最小となる要素は 1 である．<br><br>上記２つの要素（数）は 3 と 1 であり，互いに等しくない．そのため，「インデックス 0 の要素 3」の位置と「インデックス 0 以降で最小値となる要素 1」の位置を入れ替える．

```python
# ソート対象の範囲の下限を 0 とする
sort_start = 0

# インデックス 0 以降で
# [3,6,1,9,2] の要素の最小値があるインデックスを mi とする
mi = min_idx(list1, sort_start)

# 下限として選んだインデックス 0 の要素の位置と
# インデックス 0 以降で最小値となる要素の位置を入れ替える
tmp = list1[sort_start]
list1[sort_start] = list1[mi]
list1[mi] = tmp

# 元のリスト list1 = [3,6,1,9,2] に対して
# 入れ替え操作を適用した後の list1 を出力する
list1
```

## Step 2-2．リスト [1, 6, 3, 9, 2] において，<br><br>・インデックス 1 の要素は 6 である．<br>・インデックス 1 以降で最小となる要素は 2 である．<br><br>上記２つの要素（数）は 6 と 2 であり，互いに等しくない．そのため，「インデックス 1 の要素 6」の位置と「インデックス 1 以降で最小値となる要素 2」の位置を入れ替える．

```python
# ソート対象の範囲の下限を 1 とする
sort_start = 1

# インデックス 1 以降で
# [3,6,1,9,2] の要素の最小値があるインデックスを mi とする
mi = min_idx(list1, sort_start)

# 下限として選んだインデックス 1 の要素の位置と
# インデックス 1 以降で最小値となる要素の位置を入れ替える
tmp = list1[sort_start]
list1[sort_start] = list1[mi]
list1[mi] = tmp

# Step 2-1 で更新されたリスト list1 = [1,6,3,9,2] に対して
# 入れ替え操作を適用した後の list1 を出力する
list1
```

## Step 2-3．リスト [1, 2, 3, 9, 6] において，<br><br>・インデックス 2 の要素は 3 である．<br>・インデックス 2 以降で最小となる要素は 3 である．<br><br>上記２つの要素（数）はいずれも 3 であり，互いに等しい．そのため，同様の入れ替え操作を行ったところで要素の並びは変わらない．

```python
# ソート対象の範囲の下限を 2 とする
sort_start = 2

# インデックス 2 以降で
# [3,6,1,9,2] の要素の最小値があるインデックスを mi とする
mi = min_idx(list1, sort_start)

# 下限として選んだインデックス 2 の要素の位置と
# インデックス 2 以降で最小値となる要素の位置を入れ替える
tmp = list1[sort_start]
list1[sort_start] = list1[mi]
list1[mi] = tmp

# Step 2-2 で更新されたリスト list1 = [1,2,3,9,6] に対して
# 入れ替え操作を適用した後の list1 を出力する
list1
```

## Step 2-4．リスト [1, 2, 3, 9, 6] において，<br><br>・インデックス 3 の要素は 9 である．<br>・インデックス 3 以降で最小となる要素は 6 である．<br><br>上記２つの要素（数）は 9 と 6 であり，互いに等しくない．そのため，「インデックス 3 の要素 9」の位置と「インデックス 3 以降で最小値となる要素 6」の位置を入れ替える．

```python
# ソート対象の範囲の下限を 3 とする
sort_start = 3

# インデックス 3 以降で
# [3,6,1,9,2] の要素の最小値があるインデックスを mi とする
mi = min_idx(list1, sort_start)

# 下限として選んだインデックス 3 の要素の位置と
# インデックス 3 以降で最小値となる要素の位置を入れ替える
tmp = list1[sort_start]
list1[sort_start] = list1[mi]
list1[mi] = tmp

# Step 2-3 で更新されたリスト list1 = [1,2,3,9,6] に対して
# 入れ替え操作を適用した後の list1 を出力する
list1
```

## 以上のステップによって，初期のリスト list1 = [3, 6, 1, 9, 2] の要素は小さい順に整列されて，list1 = [1, 2, 3, 6, 9] となった．<br><br>【解法２＜実用的＞】<br>　上で述べた一連のソート操作を利用し，ソート対象の範囲の下限を変更することでの繰り返し処理を自動的に実施するプログラムを示す（Step 1 は解法１のものと同じ）．<br><br>Step 1．インデックス s 以降で，リスト l（エル）の要素の最小値があるインデックスを得る関数 min_idx(l, s) を定義する．

```python
def min_idx(l, s):

    # 仮にインデックス s の要素を最小値として，
    # その値を m とすると共に，インデックス s を m_idx とする
    m = l[s]  
    m_idx = s
    
    # i を大きくしながら，より小さい l[i] の値が見つかるごとに，
    # その l[i] を m に上書きすると共にインデックス i を m_idx に上書きする
    for i in range(s, len(l)):
        if m > l[i]: 
            m = l[i]
            m_idx = i
    
    # 最終的に得られる要素の最小値があるインデックスを出力する
    return m_idx
```

## Step 2．ソート対象の範囲の下限を変更することによる繰り返し処理を，while 文を使って効率よく行う．また，任意のリストに対してソートを行えるように，引数を（数を要素とする）リストとした小さい順の整列ソートを含めた関数 selection_sort_min() を定義する．

```python
def selection_sort_min(l):
    sort_start = 0
    
    while sort_start < len(l):
        mi = min_idx(l, sort_start)
        
        tmp = l[sort_start]
        l[sort_start] = l[mi]
        l[mi] = tmp
        
        sort_start += 1
```

```python
list2 = [3, 6, 1, 9, 2]
selection_sort_min(list2)
list2
```

## 注意．解法２での整列操作は，ソート対象の範囲を変えるごとに min_idx() を繰り返し実行した解法１での整列操作と同じものである．解法２における一連のアルゴリズムの方が，プログラムの文量の少なさや見易さの観点で，解法１におけるそれよりも実用的である．

## 9.2 選択ソートの実行時間（p.116～）<br><br>前節で紹介した選択ソートを使って，要素数の大きいリストのソートを行う．<br><br>例．0 から 9999 までの整数から成る 10000 個の乱数のソート<br><br>はじめに，0 から 9999 までの整数から成る 500 個の乱数を重複なく選択する．

```python
import random
list3 = random.sample(list(range(10000)), k = 500)
list3
```

## 以下のプログラムを実行することで，選択ソートにかかる時間を調べる．

```python
%%time
selection_sort_min(list3)
list3
```

## 生成させる乱数を 500 個から 5,000 個に増やして，同様に時間を測定する．

```python
list4 = random.sample(list(range(10000)), k = 5000)
list4
```

```python
%%time
selection_sort_min(list4)
list4
```

## 扱うデータ量は 10 倍であるが，処理にかかった時間は $639/6.92=92.341... \approx 100$ より，約 100 倍に増加している．これは，選択ソートのアルゴリズムを使った場合は，計算にかかる処理量がデータ量の２乗に比例して増加するためである．データ量をさらに 10 倍にして，生成させる乱数を 500 個から 50,000 個に増やした場合は，$$\Big( \frac{50000}{500} \Big)^2= 100^2=10000$$ より，データ量が 500 個の場合に比べて処理時間が約 10,000 倍にまで膨れ上がる．
