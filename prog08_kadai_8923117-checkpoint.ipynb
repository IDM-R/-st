## 問１．繰り返しを使った，$x$ を 0 から 20 まで 1 ずつ変化させて，各 $x$ の値に対する $x^4-5\sqrt{x}+1$ の値を計算し， $x$ と $x^4-5\sqrt{x}+1$ の値を表示するプログラムを作成せよ．<br><br>注意１．$5\sqrt{x}$ は<span style="color: blue; "> $\sqrt{x}$ の５倍</span>であり，$x$ の５乗根ではない．<br>注意２．講義（第８回）ノート下部の例８と同様に，$x$ の平方根 $\sqrt{x}$ を計算するために，始めの行で math モジュールをインポートすること．

## 問３．講義（第８回）ノート下部の問２に倣って，次の手順で，繰り返しを使って，「１の位が揃う形で出力されるような」九九の表（1 から 9 までの数同士の掛け算表）を作成せよ．<br><br>&emsp; &emsp; &emsp; $\bullet$ for 文 (1) で，dan という変数に 1 から 9 までのリストの値を代入する．<br>&emsp; &emsp; &emsp; &emsp; &nbsp; $\circ$ for 文 (2) を作り，gyo という変数に 1 から 9 までのリストの値を代入する．<br>&emsp; &emsp; &emsp; &emsp; &emsp;&emsp; $\triangleright$ dan と gyo の積 dan $\times$ gyo を計算し，結果を改行無しで表示する．<br>&emsp; &emsp; &emsp; &emsp; &nbsp; $\circ$ 行に対する処理が終了したら，改行する．

```python
# 問３の入力欄
for dan in range(1,10): 
    for gyo in range(1,10): 
        print("%3d" % (dan * gyo), end=" ")
    print()
```

## 問５（発展）．講義（第８回）ノート下部の問３に倣って，変数 $x,y$ を使った２重の for 文を使って，文字列の長さを 12 とした次に示す模様を表示するプログラムを作成せよ．

## $-+*-+*-+*-+$ $*$<br>$+*-+*-+*-+*$ $-$<br>$*-+*-+*-+*-$ $+$<br>$-+*-+*-+*-+$ $*$<br>$+*-+*-+*-+*$ $-$<br>$*-+*-+*-+*-$ $+$<br>$-+*-+*-+*-+$ $*$<br>$+*-+*-+*-+*$ $-$<br>$*-+*-+*-+*-$ $+$<br>$-+*-+*-+*-+$ $*$<br>$+*-+*-+*-+*$ $-$<br>$*-+*-+*-+*-$ $+$

```python
# 問５の入力欄
import math
for x in range(0,12): 
    for y in range(0,12): 
        if (x+y) % 3 == 0: 
            print("-", end=" ")
        elif(x+y)%3 ==1:
            print("+", end=" ")
        else:
            print("*", end=" ")
    print()
```
