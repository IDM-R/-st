## 復習＜4.2 for 文を使った繰り返し＞<br><br>繰り返し実行を行う別の方法として，for 文を使うこともできる．for 文では，繰り返したい要素を与えると，その要素の数だけループを繰り返す．Python の for 文は次の書式をとる．

```python
for 変数 in シーケンス:
    実行する命令1
    実行する命令2
```

## for 文の記述では次の５点に注意する．<br><br><span style="color: blue; ">(1) 繰り返しで使用する「変数」を指定する </span><br><span style="color: blue; ">(2) 変数とシーケンスの間には，キーワード「in」を挿入する</span><br>(3) if 文や while 文と同様に，シーケンスの後に半角コロン「:」が必要<br>(4) 実行する命令は必ず段付け（インデント）を与えて記述する<br>(5) 段付けされている行は，全て繰り返し実行される

## 例１－１．for 文を使った，1 から 7 までの数字それぞれを４倍した数を順に（数字の小さい順に）印字するプログラム

```python
for i in [1,2,3,4,5,6,7]:
    print(4*i)
```

## 例１－２．for 文を使った，1 から 7 までの数字それぞれを４倍した数を逆順に（数字の大きい順に）印字するプログラム<br><br>注意．要素の並び順が例１－１とは異なるので，変数 i に代入される数字の順も異なる．そのため，出力される４倍された数字の順も同じとはならない．

```python
for i in [7,6,5,4,3,2,1]: # 変数 i に，7, 6, 5, 4, 3, 2, 1 を順次代入し，以下を実行
    print(4*i)
```

## シーケンスの中身は，例１の１行目終わりのような数字でなくても構わない．

```python
for i in ["January","February","March"]:
    print("今月は",i,"です")
```

## print 文で「改行したくない」場合は，最後の引数として「end=" "」をつける．<br><br>例５．1 から 100 までの数字を「改行無しで」印字するプログラム

## 例６．range() を使った，1 から 30 までの数字の合計を求めるプログラム

```python
s=0 # 合計値を格納するための文字を s として，初期値を 0 とする
for i in range(1,31):
    s += i
print("求める合計値は",s) # 2,3 行目の繰り返し処理による合計の計算が終わった後に，最終的な合計値を出力
```

## 問１．次の手順で，繰り返しを使って，1 から 5 までの数同士の掛け算表を作成せよ．<br><br>&emsp; &emsp; &emsp; $\bullet$ for 文 (1) で，dan という変数に 1 から 5 までのリストの値を代入する．<br>&emsp; &emsp; &emsp; &emsp; &nbsp; $\circ$ for 文 (2) を作り，gyo という変数に 1 から 5 までのリストの値を代入する．<br>&emsp; &emsp; &emsp; &emsp; &emsp;&emsp; $\triangleright$ dan と gyo の積 dan $\times$ gyo を計算し，結果を改行無しで表示する．<br>&emsp; &emsp; &emsp; &emsp; &nbsp; $\circ$ 行に対する処理が終了したら，改行する．<br><br>【模範解答（問１）】

```python
for dan in range(1,6): # for 文 (1)
    for gyo in range(1,6): # for 文 (2)
        print(dan * gyo, end=" ")
    print() # gyo=5 のときの dan * gyo の出力が終わった時点で改行
```

## 注意．問１のプログラムは<span style="color: blue; ">２重の</span> for 文を含んでいる．プログラムの上から２行目に対応する<span style="color: blue; ">２段目の</span>ループ処理を記述するためには，<span style="color: blue; ">２重の</span>段付け，すなわち，（半角４文字 $\times$<span style="color: blue; ">２</span>$=$）半角８文字分の段付けを行う必要がある．

## 問２（p.54）．問１と同じ手順による繰り返しを使って，「１の位が揃う形で出力されるような」1 から 5 までの数同士の掛け算表を作成せよ．<br><br>【模範解答（問２）】

```python
for dan in range(1,6): # for 文 (1)
    for gyo in range(1,6): # for 文 (2)
        print("%3d" % (dan * gyo), end=" ") # ！！！！！dan * gyo の左右にカッコを忘れずに！！！！！
    print() # gyo=5 のときの dan * gyo の出力が終わった時点で改行
```

## 注意１．問２のプログラムにおいて，上から３行目の print() の始めにも書かれているように，演算子「%」を使うと，桁を揃えて数字を表示することができる．<br><br>print("%3d" % 数字)：３桁の数字で表示する<br>print("%4d" % 数字)：４桁の数字で表示する<br><br>ここでは，書式として，"%<span style="color: red; ">3d</span>" で<span style="color: red; ">３桁の</span>十進表示を指定している．<br><br>注意２．問２のプログラムにおいて，上から３行目の掛け算 dan * gyo の<span style="color: blue; ">左右のカッコを外してしまう</span>と，以下のような誤った出力結果が得られる（カッコの有無によって，出力結果がどのように異なるかを，各自で考察せよ）．

```python
for dan in range(1,6): # for 文 (1)
    for gyo in range(1,6): # for 文 (2)
        print("%3d" % dan * gyo, end=" ")
    print()
```

## 問３（p.54）．変数 $x,y$ を使った２重の for 文を使って，文字列の長さを 9 とした次に示す市松模様を表示するプログラムを作成せよ．

$-*-*-*-*-$<br>$*-*-*-*-*$<br>$-*-*-*-*-$<br>$*-*-*-*-*$<br>$-*-*-*-*-$<br>$*-*-*-*-*$<br>$-*-*-*-*-$<br>$*-*-*-*-*$<br>$-*-*-*-*-$

## 【模範解答（問３）】

```python
for x in range(0,9): # 1 段目のループ処理
    for y in range(0,9): # 2 段目のループ処理
        if (x+y) % 2 == 0: # (x+y) % 2：x+y を 2 で割った余り
            print("-", end=" ")
        else:
            print("*", end=" ")
    print() # y=8 のときの文字列 - あるいは * の出力が終わった時点で改行
```
