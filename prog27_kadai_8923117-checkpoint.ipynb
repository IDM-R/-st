## 問１．講義（後期，第１２回）ノートと同じく，$f$ を１変数関数として，$x_{\rm min},x_0$ を次のように定義する．<br><br>・$x_{\rm min}$：$f(x)$ が極小となるような $x$ の値 <br>・$x_0$：$x_{\rm min}$ を探すために与える $x$ の初期点<br><br>次の問に答えよ．なお，$x_0$ および勾配降下法で用いる学習率 $\alpha>0$ は，極小値が見つかる範囲内で自由に決めて良いが，$\alpha$ は小さめにとると良い．<br><br>（１）勾配降下法により，2 次関数 $f(x)=11x^2+6x+2$ の極小値（最小値）を探索せよ．

```python
# 問１（１）の入力欄
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 11*(x**2)+6*x+2

a = 0.02 # 学習率

x0 = 10 # x の初期値
plt.scatter(x0, f(x0), color="green", s=100)

h = 1e-10 # 導関数 f' を与えるための微小な値
error = 1e-8 # 収束判定用の数値
max_loop = 50 # 計算回数の最大値

num_calc = 0 # 計算回数の初期値を 0 とする
print("%3d" % num_calc,"回目の探索：x=", x0)

while(True):
    Df = (f(x0+h)-f(x0))/h
    x1 = x0 - a*Df # 根の近似値の更新
    num_calc += 1 # 計算回数のカウントアップ
    print("%3d" % num_calc, "回目の探索：x=", x1)
    plt.plot(x1, f(x1), marker='.', markersize=10, color="black") # 根の近似値とそのときの f の値を点で表示

    
    # 終了条件１：更新前後で x の増分（の絶対値）が十分小さくなったときに，探索を終了する
    if(abs(x1-x0) <= error):
        print()
        print("x_min の探索移動量：small enough (<=", error, ").")
        plt.plot(x1, f(x1), marker='.', markersize=20, color="blue") # 根の近似値とそのときの f の値を点で表示
        print("x_min=", x1, "，f(x_min)=", f(x1))
        break
        
    # 終了条件２：max_loop 回分の計算を終えた場合でも，探索を終了する
    if(max_loop<=num_calc):
        print("The number of loops reach a maximum value.")
        break
    
    x0 = x1 # 根の近似値を更新

### グラフ描画 ###
x = np.linspace(-15, 20, 201)
plt.plot(x, f(x))

### オプション ###
plt.xlabel("$x$") # x 軸の名前
plt.ylabel("$f(x)$") # y 軸の名前
plt.grid()
```

## 【問３の入力欄】<br><br>（１）「２変数」関数 $f(x,y)$ の最小化問題に対する勾配降下法<br><br>勾配降下法は関数の微分値から傾きに従って極小値を求める手法である。一変数の場合とは異なり、グラフ上で曲線ではなく曲面で表される。<br><br> $f$ が極小となるような $x,y$ をそれぞれ $x_{\rm min},y_{\rm min}$ x}f,\frac{\partial}{\partial y}f$ の値は大きくなり，横向き移動量も大きくなる．<br>・$x,y$ が $x_{\rm min},y_{\rm min}$ に近い場合，$\frac{\partial}{\partial x}f,\frac{\partial}{\partial y}f$ の値は小さくなり，横向き移動量も小さくなる．<br><br>$x_{\rm min},y_{\rm min}$ を探索するプログラムを下のセルで実行した。<br><br>【注意】 $\alpha$ を極端に大きく・小さくとると，1変数の勾配降下法同様に次のメリットやデメリットが生じる．<br><br>$\alpha$ が極端に大きいとき：<br>　少ない計算回数で $x_ny_n$ は $x_{\rm min}y_{\rm min}$ がとるための計算回数が多くなるために計算コストがかかり，処理時間が極端に長くなることもある．<br><br>また、極小値や停留点など、局所会に収束してしまう可能性がある。<br><br>さらに、2変数の場合には一方の変数の最小であるがもう一方の変数では最大となる鞍点付近で留まったりと、両変数の最小値に収束しない場合もあるかもしれない。<br><br>このような勾配降下法はAIの学習方法であるディープラーニングにおいても用いられている。ディープラーニングで用いられる、最小値問題に対処する最急降下法はほかにもあり、それぞれメリットとデメリットが存在する。将来、計算コストが少なく正確に最小値問題を解決する手法の開発に期待したい。<br><br>（２）回帰分析の現実社会への応用<br><br>回帰分析は統計学の手法の一つであり、経済分野をはじめ、医療分野やビジネス、教育まで現実社会においてさまざまな応用がある。<br><br>経済学とビジネス分野では、価格予測や収益予測に用いられている。商品やサービスの価格を予測するために回帰分析が使用され、需要と供給、広告出稿、競合他社の価格などの変数を用いて価格設定を最適化している。また、過去のデータから未来の収益を予測するために回帰分析が利用され、企業の戦略や予算の策定を行っている。<br><br>医療と生物統計学においては、患者の臨床データや生体検体から得られる情報をもとに、病気の進行を予測するためなどに回帰分析が使用され、治療計画や患者の予後の評価を行う。<br><br>さらに、マーケティングにも活用されており、広告支出と売上の関係を分析し、どの広告が最も効果的であるかを評価する広告の効果測定や企業のマーケティング活動や競合他社の動向をもとに、市場シェアの予測をする際に回帰分析が用いられる。<br><br>次に教育、学習分野においては、学生の学習成績に影響を与える要因を分析し、将来の成績を予測するために回帰分析が応用され、教育政策や個別の学習サポートが検討できる。<br><br>最後に労働市場と人事管理においても活用されているので紹介する。この分野では、企業が従業員の給与を適正に設定するために、経験、スキル、役職などの要因を考慮して給与を予測したり、失業率や人口統計などの変数を用いて、将来の雇用動向を予測したりするために回帰分析が利用される。<br><br>以上から、回帰分析はこれらの分野で様々な要因の影響を定量的に評価し、未来の状況を予測する際に有用なツールであるといえる。しかし、これらは数値として与えられているものからしか分析できない。私たちの課題は現実の課題の要因を発見し、定量化することである。私たちはAIによる回帰分析の技術的進歩の恩恵をうけつつ、社会にのさばっている様々な課題の原因を定量化し、課題解決をしていくべきだろう。


```python

def f(x, y):
    z = x ** 2 + y ** 2
    return z
# 基準関数の微分
def df(x, y):
    dzdx = 2 * x
    dzdy = 2 * y
    dz = np.array([dzdx, dzdy])
    return dz
 
# 勾配降下法に必要なパラメータ
eta = 0.1                           # 学習率
max_iteration = 100            # 最大反復回数
x0 = -5                           # 初期値x0
y0 = 5                            # 初期値y0
x_pred = [x0]                       # 描画用x0軌跡リスト(初期値をプリセット)
y_pred = [y0]                       # 描画用y0軌跡リスト(初期値をプリセット)
 
# 最大反復回数まで計算する
for i in range(max_iteration):
    x0, y0 = np.array([x0, y0]) - eta * df(x0, y0)          # 勾配降下法
    x_pred.append(x0)               # x0の軌跡をリストに追加
    y_pred.append(y0)               # y0の軌跡をリストに追加
    print(i, x0, y0)
 
```

```python
x_pred = np.array(x_pred)           # 描画用にx0をnumpy配列変換
y_pred = np.array(y_pred)           # 描画用にx0をnumpy配列変換
z_pred = f(x_pred, y_pred)          # 軌跡のz値を計算
 
# 基準関数の表示用
x = np.arange(-10, 11, 2)
y = np.arange(-10, 11, 2)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
 
# ここからグラフ描画----------------------------------------------------------------
# フォントの種類とサイズを設定する。
plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'Times New Roman'
 
#  グラフの入れ物を用意する。
fig = plt.figure()
ax1 = fig.add_subplot(projection='3d')
 
# 軸のラベルを設定する。
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')
 
# データプロットする。
ax1.plot_wireframe(X, Y, Z, label='f(x, y)')
ax1.scatter(x_pred, y_pred, z_pred, label='gd', color='red', s=50)
 
# グラフを表示する。
plt.show()
plt.close()
```
